```python
import pandas as pd
```

```python
def read_tntp_network(file_path):
    with open(file_path, 'r') as f:
        lines = f.readlines()

    # Find where the header line starts (after metadata)
    header_idx = 7
    for i, line in enumerate(lines):
        if line.strip().startswith('From'):
            header_idx = i
            break

    # Read the file into a DataFrame
    df = pd.read_csv(
        file_path,
        sep=r'\s+',                   # whitespace separator
        skiprows=header_idx,          # skip metadata lines
        comment='~',                  # ignore comment lines
        names = ['From', 'To', 'Capacity', 'Length', 'FreeFlowTime', 'B', 'Power', 'SpeedLimit', 'CriticalSpeed', 'Lanes', 'SemiColon'],
        engine='python'
    )


    return df

# Example usage:
df_network = read_tntp_network("../datasets/Goldcoast_network_2016_01.tntp")
df_network= df_network.drop("SemiColon",axis=1)
df_network


```

```python
df_network.to_csv("GoldCoast_network.csv", index=False)
```
